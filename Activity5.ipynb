```python
import pandas as pd 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
df = pd.read_csv('Association_Rule.csv')
df
```

```python

data =  df.iloc[:, [0,1,2,12]]
#data = data.insert(0, 'Instance', range(len(data))) error
data
```

```python
#data = data.drop(['Combination'], axis=1)
data['Transaction_ID'] = range(880, 880+len(data["Credit Hist"]))
#data = data.rename(columns={"Transaction": "Transaction_ID"})
data
```

```python
#Rearrange
data = data[["Transaction_ID", "Checking Acct", "Credit Hist", "Purpose", "Credit Standing"]]
data
```

```python
combine_columns = lambda row: f"{row['Checking Acct']} {row['Credit Hist']} {row['Purpose']}"
data['Description'] = data.apply(combine_columns, axis=1)
data
```

```python
final_data =  data.iloc[:, [0,5]]
final_data['CreditStandingBool'] = data['Credit Standing'].apply(lambda x: 0 if x == 'Bad' else 1)
final_data
```

```python
# Stripping extra spaces in the description
final_data['Description'] = final_data['Description'].str.strip()

final_data
```

```python
basket = (final_data.groupby(['Transaction_ID', 'Description'])['CreditStandingBool']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Transaction_ID'))
basket
```

```python
#Convert all positive number to one
#Convert all less than zero to 0
def encode_units(x):
    if x <= 0:
        return 0
    if x>= 1:
        return 1
basket_sets =basket.applymap(encode_units)#Apply a function to a Dataframe elementwise. This method applies a function that accepts and returns a scalar to every element of a DataFrame.
#basket_sets.drop('POSTAGE', inplace=True, axis=1) Error: POSTAGE, Solution: remove this line of code
#basket_sets.dropna(inplace=True, axis=0) #Error: Found nan, Solution: This line of code
basket_sets
```

```python
#Generate frequent itemsets with support at least 7%
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
